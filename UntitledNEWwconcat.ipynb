```python
#######THIS CONCATENATES ALL BestFit FILES#######

#######CREATES DATAFRAME FROM FIRST BESTFITNESS FILE FIRST#######
#######change the text inside the double quotes#######
runFitness1 = pd.read_csv("/Users/ritchiemaegamot/Documents/upmin/CHIO w Chaotic Maps/CC1/CC1run1to17/run1/CC1_BestFit.csv",header=None)

#######CONCATENATE THE REST OF THE BESTFITNESS FILES#######
#######THIS ADDS RUN 2 TO 17, THIS IS DEPENDING ON HOW THE FOLDER WAS SET UP#######

#######change the range values#######
#######if the run folders go from 2 to 17, then format is range(start, end+1, 1)#######
for x in range(2, 18, 1): 
    #######change the text inside the double quotes#######    
    directory = "/Users/ritchiemaegamot/Documents/upmin/CHIO w Chaotic Maps/CC1/CC1run1to17/"
    addPath = "run" + str(x) + "/CC1_BestFit.csv"
    fullPath = directory +  addPath

    pandaName ="runFitness" + str(x)
    pandaName = pd.read_csv(fullPath,header=None)
    
    runFitness = pd.concat([runFitness1, pandaName], axis=1)
    runFitness1 = runFitness.copy()

####### CONVERTS THE CONCATENATED DATAFRAMES TO CSV#######
runFitness.to_csv('CC1_BestFit_all.csv',header=None, index= False)

#######CONCATENATE THE REST OF THE BESTFITNESS FILES#######
#######THIS ADDS RUN 18 TO 30, THIS IS DEPENDING ON HOW THE FOLDER WAS SET UP#######

#######change the range values#######
#######if the run folders go from 18 to 30, then format is range(start, end+1, 1)#######
for x in range(18, 31, 1): 
    #######change the text inside the double quotes#######
    directory = "/Users/ritchiemaegamot/Documents/upmin/CHIO w Chaotic Maps/CC1/CC1run18to30/"
    addPath = "run" + str(x) + "/CC1_BestFit.csv"
    fullPath = directory +  addPath

    pandaName ="runFitness" + str(x)
    pandaName = pd.read_csv(fullPath,header=None)
    
    runFitness = pd.concat([runFitness1, pandaName], axis=1)
    runFitness1 = runFitness.copy()
    
# print(runFitness)

runFitness.to_csv('CC1_BestFit_all.csv',header=None, index= False)
runFitness = pd.read_csv("CC1_BestFit_all.csv",header=None)
```

```python

num_iters = 30001
x = np.linspace(0, num_iters,num_iters)

for i in range(0,len(runFitness.iloc[0])):
    plt.plot(x,runFitness[i],'g-')


plt.title('Convergence Graph of CHIO 6-center \n problem  applied on Davao City, Philippines dataset.')
plt.xlabel('Iteration Number')
plt.text(20000,12000,"Best   = {:05.3f}".format(min(runFitness.iloc[30000]))) 
plt.text(20000,11700,"Worst  = {:05.3f}".format(max(runFitness.iloc[30000]))) 
plt.ylabel('Maximum Distance')

plt.savefig('runFitnessP6_CC1.jpg',dpi=300,bbox_inches='tight')
```
